# Using `numpy.vander` to calculate numerical polynomials

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-15,15,400)
```

```python
coeffs = np.array([0.0333, 0,-2.50, 0])
# if we use default state of vander, this starts with cubic coefficient
# and ends with the 0th power coefficient.
```

```python
A = np.vander(x,4)
```

```python
y = np.dot(A, coeffs)
```

```python
plt.plot(x,y)
plt.grid()
```

```python
X = np.array([x + np.random.normal(scale=0.5) for x in range(-15,15)])
#spaced about 1 apart with some jitter
```

```python
Y = np.dot(np.vander(X,4), coeffs) 
Y += np.random.normal(scale = 5, size = np.shape(Y))
```

```python
plt.plot(x,y)
plt.scatter(X,Y)
plt.grid()
```
